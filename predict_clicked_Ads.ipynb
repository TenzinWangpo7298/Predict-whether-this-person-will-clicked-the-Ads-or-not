# Logistic Regression

```python
#Split the data into training set and testing set using train_test_split
from sklearn.model_selection import train_test_split
```

```python
#since our objective is to find the probability of internet user clicked the ads 
X = adver[['Daily Time Spent on Site', 'Age','Area Income','Daily Internet Usage', 'Male']]
y = adver["Clicked on Ad"]
```

** Train and fit a logistic regression model on the training set.**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit = LogisticRegression()
```

```python
logit.fit(X_train,y_train)
```

## Predictions and Evaluations

```python
results = logit.predict(X_test)
```

```python
#Create a classification report for the model.
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,results))
```
